### Feature Engineering

In this section we will do the feature engineering on dataset. Basically we will perform some operation and change the dataset according to our need.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('train.csv')
dataset.head()
```

**Missing Value**

```python
categorical = [feature for feature in dataset.columns if dataset[feature].dtype=="O" and dataset[feature].isnull().sum()>0]
for feature in categorical:
    print(feature,"Missing value % {}".format(dataset[feature].isnull().sum()/len(dataset)))
```

```python
# In order to handle missing value in categorical features we simply replace NaN with some different label and treat them as 
#valid value 
def handle_cat_missing(dataset,categorical):
    data = dataset.copy()
    data[categorical]=data[categorical].fillna("TempVal")
    return data

dataset = handle_cat_missing(dataset,categorical)

#Now cross check if it filled or not
dataset[categorical].isnull().sum()


```

```python
dataset[categorical].head(50)
```

**Handle numerical missing value**

```python
numerical = [feature for feature in dataset.columns if dataset[feature].dtype !="O" and dataset[feature].isnull().sum()>0]
numerical
```

```python

for feature in numerical:
    # Here we wll fill the null value with median since it has outliers
    median_val = dataset[feature].median()
    #create separate feature for nan val
    dataset[feature + "_na"]=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_val,inplace= True)
    
na=[feature for feature in dataset.columns if '_na' in feature]
na

dataset[numerical].isnull().sum()
```

```python
dataset.head()
```

```python
# Now handle Temporal feature

for temporal in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[temporal]=dataset['YrSold'] - dataset[temporal]
    
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']].head()
```

**Handle skewness of Numerical contenous variable**

We have seen already that some of the numerical features having skeweness of the data
So we need to perform logrithmic transformatin.

```python
numerical_contenous= [ 'LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in numerical_contenous:
    dataset[feature] = np.log(dataset[feature])

dataset.head()    

```

**Handle Rare Variable (Less used variable)**

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    #print (temp)
    temp_df=temp[temp>0.01].index
    #print(temp_df)
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head()
```

**Feature Scalling**

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[scaling_feature])
```

```python
scaler.transform(dataset[scaling_feature])
```

```python
#transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[scaling_feature]), columns=scaling_feature)],
                    axis=1)
```

```python
data.to_csv('X_train.csv',index=False)
```
